# 3. Analysis of Data:
To address these objectives, various statistical tests are performed, each with specific goals:

### Two-Sample Independent t-Test:
- Compares the means of transaction values between two countries.
- **Null hypothesis (H₀):** The means of transaction values between the two countries are equal.
- **Alternative hypothesis (H₁):** The means of transaction values differ.
- A significant p-value (< 0.05) indicates a difference between the two countries.

### Analysis of Variance (ANOVA):
- Used when comparing transaction values across more than two countries.
- **Null hypothesis (H₀):** All countries have the same mean transaction value.
- **Alternative hypothesis (H₁):** At least one country has a different mean.
- Post-hoc tests (e.g., Tukey's HSD) can be used to identify which countries differ.

### Chi-Square Test of Independence:
- Tests for an association between two categorical variables, such as country and product category.
- **Null hypothesis (H₀):** Country and product category are independent.
- **Alternative hypothesis (H₁):** There is a significant association between country and product category.

### Correlation Analysis:
- Examines the relationship between two continuous variables (e.g., transaction value and time).
- **Pearson Correlation:** Measures linear relationships.
- **Spearman Correlation:** Measures monotonic relationships (non-linear trends).

### Regression Analysis (Optional):
- If multiple predictors (e.g., product type, country) are available, a regression model can be built to predict transaction value.
- **Multiple Linear Regression:** Helps quantify how each factor influences transaction value.

# 4. Observations and Findings:
The results from statistical tests yield several important insights:

- **T-Test Findings:** A significant t-test result indicates that transaction values vary significantly between two countries. This could highlight differences in purchasing power or regional economic conditions.
- **ANOVA Findings:** If the ANOVA result is significant, it shows that not all countries have the same average transaction value. Post-hoc tests can identify which countries stand out.
- **Chi-Square Findings:** A significant Chi-Square result indicates that product categories and countries are associated, revealing regional preferences for certain products.
- **Correlation Findings:** Significant correlations between transaction values and other continuous variables suggest possible predictors of transaction behavior, such as time or quantity.
- **Regression Findings (if applicable):** A well-fitting regression model helps predict future transaction values based on variables like country and product category, aiding strategic decision-making.
